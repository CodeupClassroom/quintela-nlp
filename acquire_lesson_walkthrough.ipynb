## NY Times Website

```python
response = requests.get('https://www.nytimes.com/')
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
soup.find_all('h3')
```

```python
soup.find_all('h3', class_='indicate-hover')
```

```python
[el.text for el in soup.find_all('h3', class_='indicate-hover')]
```

```python
soup.find_all('a')
```

```python
article_links = soup.find_all('a', class_='css-9mylee')
article_links
```

```python
article_links[0]['href']
```

```python
[el['href'] for el in article_links]
```

```python
for element in article_links:
    print(element['href'])
```

```python
links_list = []

for element in article_links:
    links_list.append(element['href'])
```

```python
links_list
```

```python
url = links_list[0]

requests.get(url)
```
